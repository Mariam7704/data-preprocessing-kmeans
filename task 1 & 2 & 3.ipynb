```python
dataset.rename(columns={'name': 'title'}, inplace=True)
dataset.head(5)
```

```python
dataset[['title', 'year']]
```

```python
excluded_ratings = ['G', 'Unrated', 'NC-17', 'TV-MA', 'TV-PG', 'X', 'Approved', 'TV-14']
dataset = dataset[~dataset['rating'].isin(excluded_ratings)]
```

```python
excluded_genres = ['Biography', 'Animation', 'Fantasy', 'Mystery', 'Family', 'Musical', 'Music', 'History', 'Sport','Thriller','Sci-Fi','Romance', 'Western']
dataset = dataset[~dataset['genre'].isin(excluded_genres)]
```

Data cleaning and Preprocessing

```python
dataset.isnull()
```

Dealing with missing data

```python
dataset.isna().sum()
```

```python
nu_sum=dataset.isna().sum()
nu_sum[nu_sum>0]
```

```python
sns.heatmap(dataset.isnull())
```

```python
budget_median = dataset['budget'].median()
gross_median = dataset['gross'].median()

dataset['budget']=dataset['budget'].fillna(budget_median)
dataset['gross']=dataset['gross'].fillna(gross_median)
sns.heatmap(dataset.isnull())
```

Convert Categorical Features

```python
dataset['genre'].value_counts()
```

```python
dataset['genre'].value_counts().plot.pie(autopct='%1.0f%%')
```

```python
dataset['rating'].value_counts()
```

```python
dataset = dataset[(dataset['rating'] != 'Not Rated') & (dataset['rating'] != 'Unrated')]
dataset['rating'].value_counts().plot.pie(autopct='%1.0f%%')
```

```python
rating_dummies= pd.get_dummies(dataset['rating'], dtype='int')
rating_dummies
```

```python
dataset = pd.concat([dataset.reset_index(drop=True), rating_dummies.reset_index(drop=True)], axis=1)
dataset.head(2)
```

```python
genre_dummies = pd.get_dummies(dataset['genre'], dtype='int')
genre_dummies
```

```python
dataset = pd.concat([dataset.reset_index(drop=True), genre_dummies.reset_index(drop=True)], axis=1)
dataset.head(2)
```

```python
dataset.drop(['company','country','star','writer','director','released'],axis=1,inplace=True)
dataset.info()
```

Dealing with outliers

```python
dataset['gross'].plot.hist(bins=10)
```

```python
dataset['gross'].plot.box()
```

```python
q1 = dataset['budget'].quantile(0.25)
q3 = dataset['budget'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
dataset = dataset[(dataset['budget'] >= lower_bound) & (dataset['budget'] <= upper_bound)]

dataset['budget'].plot.hist(bins=10)
```
